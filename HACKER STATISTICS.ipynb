# DataCamp Project

The project is to iterate through a number steps to get to the top of a 60 stairs building
using a random die 
if the die falls on a number less than 2, you go down a step of one
if the die falls between 3 and 5 you go up one
else you go up on a random number between 1 and 6

```python
import numpy as np
```

```python
np.random.seed(123)
```

```python
all_walks = []
```

#simulate 500 steps

```python
for i in range(500):
    random_walk = [0] #starting at the ground floor
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step-1) #making sure the steps do not go below 0
        elif dice <=5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        #checking for clumsiness
        if np.random.rand()<=0.001:
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
```

```python
np_all_walks = np.array(all_walks)
```

```python
print(np_all_walks)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np_all_walks)
plt.show()
```
